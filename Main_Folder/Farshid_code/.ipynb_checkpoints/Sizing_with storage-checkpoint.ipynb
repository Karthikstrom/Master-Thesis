```python
Effi_out= 1  
Effi_in=1
df_whole.loc[0,'P_available_old'] = 5
df_whole.loc[0,'P_available_new']=5
df_whole.loc[0,'P_dis'] = 0
df_whole.loc[0,'P_char'] = 0
df_whole['P_s_max'] = 3
df_whole['P_dump']=0
for i in range (len(df_whole)):
 if (df_whole.loc[i,'P_output(Kw/h)'] < df_whole.loc[i,'Kw/h']) and ((df_whole.loc[i,'Kw/h'] - df_whole.loc[i,'P_output(Kw/h)']) < df_whole.loc[i,'P_available_new']):
        df_whole.loc[i, 'Arbitrage'] = 0
        df_whole.loc[i, 'P_dis'] = (df_whole.loc[i,'Kw/h'] - df_whole.loc[i,'P_output(Kw/h)'])
        df_whole.loc[i, 'P_char'] = 0
        df_whole.loc[i+1, 'P_available_old'] = (df_whole.loc[i, 'P_available_new'])
        df_whole.loc[i+1, 'P_available_new'] = (df_whole.loc[i+1,'P_available_old'] + (df_whole.loc[i,'P_char']*Effi_in) - (df_whole.loc[i,'P_dis']*Effi_out))
 elif (df_whole.loc[i,'P_output(Kw/h)'] < df_whole.loc[i,'Kw/h']) and ((df_whole.loc[i,'Kw/h'] - df_whole.loc[i,'P_output(Kw/h)']) > df_whole.loc[i,'P_available_new'] ):
        df_whole.loc[i, 'Arbitrage'] = df_whole.loc[i,'Kw/h'] - df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'P_available_new']
        df_whole.loc[i, 'P_dis'] = df_whole.loc[i,'P_available_new']
        df_whole.loc[i, 'P_char'] = 0
        df_whole.loc[i+1, 'P_available_old'] = df_whole.loc[i, 'P_available_new']  
        df_whole.loc[i+1, 'P_available_new'] = df_whole.loc[i+1,'P_available_old'] + (df_whole.loc[i,'P_char']*Effi_in) - (df_whole.loc[i,'P_dis']*Effi_out)
 elif (df_whole.loc[i,'P_output(Kw/h)'] > df_whole.loc[i,'Kw/h']) and ((df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h']) < (P_max_battery-df_whole.loc[i,'P_available_new'])):
        df_whole.loc[i, 'Arbitrage'] = 0
        df_whole.loc[i, 'P_dis'] = 0
        df_whole.loc[i, 'P_char'] = df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h']
        df_whole.loc[i+1, 'P_available_old'] = df_whole.loc[i, 'P_available_new']
        df_whole.loc[i+1, 'P_available_new'] = df_whole.loc[i+1,'P_available_old'] + (df_whole.loc[i,'P_char']*Effi_in) - (df_whole.loc[i,'P_dis']*Effi_out)       
 elif (df_whole.loc[i,'P_output(Kw/h)'] > df_whole.loc[i,'Kw/h']) and ((df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h']) > (P_max_battery-df_whole.loc[i,'P_available_new'])) and ((df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h'] -(P_max_battery-df_whole.loc[i,'P_available_new'])) < df_whole.loc[i,'P_s_max']):
        df_whole.loc[i, 'Arbitrage'] = -1*(df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h'] - (P_max_battery-df_whole.loc[i,'P_available_new']))
        df_whole.loc[i, 'P_dis'] = 0
        df_whole.loc[i, 'P_char'] = P_max_battery-df_whole.loc[i,'P_available_new']
        df_whole.loc[i+1, 'P_available_old'] = df_whole.loc[i, 'P_available_new']
        df_whole.loc[i+1, 'P_available_new'] = df_whole.loc[i+1,'P_available_old'] + (df_whole.loc[i,'P_char']*Effi_in) - (df_whole.loc[i,'P_dis']*Effi_out)
 elif (df_whole.loc[i,'P_output(Kw/h)'] > df_whole.loc[i,'Kw/h']) & ((df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h']) > (P_max_battery-df_whole.loc[i,'P_available_new'])) & ((df_whole.loc[i,'P_output(Kw/h)'] - df_whole.loc[i,'Kw/h'] - (P_max_battery-df_whole.loc[i,'P_available_new'])) > df_whole.loc[i,'P_s_max']):
        df_whole.loc[i, 'Arbitrage'] = -1*(df_whole.loc[i,'P_s_max'])
        df_whole.loc[i, 'P_dis'] = 0
        df_whole.loc[i, 'P_char'] = P_max_battery-df_whole.loc[i,'P_available_new']
        df_whole.loc[i+1, 'P_available_old'] = df_whole.loc[i, 'P_available_new']
        df_whole.loc[i+1, 'P_available_new'] = df_whole.loc[i+1,'P_available_old'] + (df_whole.loc[i,'P_char']*Effi_in) - (df_whole.loc[i,'P_dis']*Effi_out)
        df_whole.loc[i, 'P_dump'] = (df_whole.loc[i, 'P_output(Kw/h)'] - df_whole.loc[i, 'Kw/h'] - (df_whole.loc[i, 'P_max_battery'] - df_whole.loc[i, 'P_available_new']) - df_whole.loc[i, 'P_s_max'])
```
